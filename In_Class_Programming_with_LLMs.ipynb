```python
from textblob import TextBlob

def analyze_sentiment(text: str) -> str:
    """Return Positive / Negative / Neutral for the given text."""
    polarity = TextBlob(text).sentiment.polarity  # between -1.0 and 1.0
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Example inputs
examples = [
    "I love this product!",
    "This is the worst movie ever.",
    "It's good week.",
    "It's a week.",
    "It's the worst week.",
    "Absolutely fantastic!",
    "I don't like it at all its worst .",
    "It's a sunny day."
]

for e in examples:
    print(f"{e} ‚Üí {analyze_sentiment(e)}")
```

```python
from google.colab import files
import pandas as pd
from textblob import TextBlob
import chardet

# 1) Upload the file
uploaded = files.upload()
filename = next(iter(uploaded.keys()))

# 2) Detect encoding
with open(filename, 'rb') as f:
    rawdata = f.read(100000)   # read first 100k bytes
result = chardet.detect(rawdata)
encoding_used = result['encoding']
print("üìÇ Detected encoding:", encoding_used)

# 3) Load dataset with detected encoding (NO 'errors' param)
df = pd.read_csv(filename, encoding=encoding_used)

print("‚úÖ File loaded successfully")
print("üìä Columns in dataset:", df.columns.tolist())
display(df.head())

# 4) Define sentiment analyzer
def analyze_sentiment(text, neg_thresh=-0.1, pos_thresh=0.1):
    polarity = TextBlob(str(text)).sentiment.polarity
    if polarity > pos_thresh:
        return "Positive"
    elif polarity < neg_thresh:
        return "Negative"
    else:
        return "Neutral"

# 5) Pick the text column (auto-detect common names)
text_column = None
for col in ["text", "review", "sentence", "content", "tweet"]:
    if col in df.columns:
        text_column = col
        break

if text_column is None:
    raise ValueError("‚ùå Could not find a text column. Please check your dataset columns.")

print(f"üìù Using column: {text_column}")

# 6) Apply sentiment analysis
df['predicted_sentiment'] = df[text_column].apply(analyze_sentiment)

# 7) Show results
display(df[[text_column, 'predicted_sentiment']].head())

# 8) Save results for download
df.to_csv("sentiment_results.csv", index=False)
files.download("sentiment_results.csv")
```
